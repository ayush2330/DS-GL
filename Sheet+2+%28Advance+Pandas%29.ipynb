### Concatination of series and array

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ser1 = pd.Series(np.random.randint(10,30,10))
ser2 = pd.Series(np.random.randint(10,30,10))
ser3 = pd.Series(np.random.randint(10,30,10))
```

```python
# Concatenate ser1 and ser2
pd.concat([ser1,ser2],ignore_index=True)
```

```python
# concatenate ser1 and ser2 with hierarchial indexing
s=pd.concat([ser1,ser2],keys=['Cls-A','Cls-B'])
s
```

```python
s['Cls-B'][0]
```

```python
pd.concat([ser1,ser2,ser3],axis=1,keys=['Cls-A','Cls-B','Cls-C'])
```

```python
# Concatenate the series with regular indexes (ignoring origional indexes)
pd.concat([ser1,ser2,ser3],ignore_index=True)
```

### Merge the dataframe

```python
df1 = pd.read_excel('data.xlsx',sheet_name=0)
df2 = pd.read_excel(r'C:\Users\bittu\Desktop\GL\Python Offline\NPV\InClass\data.xlsx',sheet_name=1)
df3 = pd.read_excel('data.xlsx',sheet_name=2)
```

```python
# Inner Join dataframe df1 and df3 using merge (default)
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True)
```

```python
# left Join dataframe df1 and df3 using merge
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True,how='left')
```

```python
# Right Join dataframe df1 and df3 using merge
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True,how='right')
```

```python
# Outer Join dataframe df1 and df2 using merge
df1.merge(df3,on='Emp ID',indicator=True,how='outer')
```
