### LINECHART

```python
x = np.array([1,2,3,4,5,6])
y = np.array([2,4,5,6,8,11])
```

```python
plt.plot(x,y,color='grey',ls='--',lw=3,marker='<',ms=12)
plt.title('LINE CHART')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
data = [1,2,3,4]
vivo = [80000,78000,87000,95000]
samsung = [45000,45000,60333,54000]
oppo = [75000,44000,58000,40888]
micro = [47000,55000,78000,65700]
```

```python
plt.figure(figsize=(7,5))   # to change the size of a particular figure
plt.title('Comparision')
plt.plot(data,vivo,marker='o',label='Vivo')
plt.plot(data,samsung,marker='o',label='Samsung')
plt.plot(data,oppo,marker='o',label='Oppo')
plt.plot(data,micro,marker='o',label='Micro')
plt.legend(loc='upper left')   # to show the labels
plt.xticks(data)    # to change range in x axis
plt.yticks(range(30000,120000,10000))   # to change range in y axis
plt.label()
plt.show()
```

### SCATTER PLOT

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
# Make a scatter plot between sepal length and sepal width
plt.scatter(iris['sepal_length'],iris['sepal_width'],color='tomato',s=20,marker='*')
plt.title('IRIS')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Make a multiple scatter plot between sepal lenght sepal width, sepal length an pela length
# Make a scatter plot between sepal length and sepal width
plt.scatter(iris['sepal_length'],iris['sepal_width'],s=20,marker='*',label='Sepal')
plt.scatter(iris['sepal_length'],iris['petal_length'],s=20,marker='x',label='Petal')
plt.legend()
plt.title('IRIS')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
#subplot -> to display multiple plots on the same display area
plt.figure(figsize=(5,3))
plt.suptitle('IRIS')   # to give title for both the charts

plt.subplot(1,2,1)
plt.scatter(iris['sepal_length'],iris['sepal_width'],s=20,marker='*',label='Sepal',color='green')
plt.title('IRIS')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.subplot(1,2,2)
plt.scatter(iris['sepal_length'],iris['petal_length'],s=20,marker='x',label='Petal')
plt.title('IRIS')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.tight_layout()
plt.show()
```

#### BAR PLOT

```python
amount = [2500,3700,3000,4200,2200]
cust = ['Ajay','Suresh','Kamla','Neha','Vijay']
```

```python
plt.bar(x=cust,height=amount,width=0.8,color=['red','blue','green','magenta','orange'])
#by default width is 0.8
for x,y in zip(range(5),amount):   # to print values for all the graphs
	plt.text(x-0.2,y+100,str(y))
plt.yticks(range(0,5001,1000))
plt.grid()
plt.figure(figsize=(5,5))


plt.show()
```

```python
# make a vertical barplot b/w amount and cust
```

```python
# make a horizontal bar plot b/w amount and cust
plt.barh(y=cust,width=amount,height=0.5,color=['red','blue','green','magenta','orange'])  #to print bar in horizontal bar
plt.show()
```

#### GROUPED BAR PLOT

```python
itp = np.random.randint(30,50,5)
npv = np.random.randint(30,50,5)
```

```python
x = np.arange(1,6)
```

```python
itp
```

```python
npv
```

```python
# Make a grouped bar plot between python npv and x
#ek ke baju me dusra
plt.bar(x=x-0.2,height=itp,width=0.4,color='orchid',label='ITP')
plt.bar(x=x+0.2,height=npv,width=0.4,color='aqua',label='NPV')  #only valid if x is a no
plt.xticks(x,cust)   #to print names we have to use x with category
plt.legend(loc='upper left')
plt.show()
```

```python
# Make a stacked bar plot between python npv and x
#ek ke upar ek graph
plt.bar(x=cust,height=itp,color='red',label='ITP')
plt.bar(x=cust,height=npv,bottom=itp,color='blue',label='NPV')
plt.legend()
plt.show()
```

#### PIE PLOT

```python
countries = ['USA','France','Germany','India','Russia']
count = [63.76,34.54,45.11,57.43,60.54]
```

```python
# Make a pie plot
plt.pie(x=count,labels=countries,radius=1,colors=['teal','lavender','blueviolet','orangered','lime'],autopct='%0.2f%%',
       pctdistance=0.8,labeldistance=1.2,startangle=90,counterclock=False,shadow=True,explode=[0.1,0,0,0,0])
#autopct is used to give the precentage=
#pct distance is percentage distance and label distance to give distance in label
plt.show()
```

```python
# make a doughnut chart
plt.pie(x=count,labels=countries,radius=1.3,colors=['teal','lavender','blueviolet','orangered','lime'],autopct='%0.2f%%',
       pctdistance=0.8,labeldistance=1.2,startangle=90,counterclock=False,shadow=True,explode=[0.2,0.2,0,0,0])
plt.pie(x=[1],colors=['w'],radius=0.7)
plt.show()
```

### SEABORN

#### STRIP PLOT

```python
data = sns.load_dataset('titanic')
data
```

```python
# Make a strip plot between age and sex
sns.stripplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
# Make a strip plot between age and sex and survived
sns.stripplot(x=data['sex'],y=data['age'],hue=data['survived'])
plt.show()
```

#### VIOLIN PLOT

```python
# Make a violin plot between age and sex
sns.violinplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
# Make a violin plot between age and sex and survived
sns.violinplot(x=data['sex'],y=data['age'],hue=data['survived'])
```

#### SWARM PLOT

```python
# Make a swarm plot between age and sex
sns.swarmplot(x=data['sex'],y=data['age'])
plt.show()
```

```python
# Make a swarm plot between age and sex and survived
sns.swarmplot(x=data['sex'],y=data['age'],hue=data['survived'],palette=['orchid','aqua'])
```
